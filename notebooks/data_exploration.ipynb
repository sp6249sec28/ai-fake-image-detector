```python
import os
import matplotlib.pyplot as plt
import cv2

real_path = "C:/Users/Administrator/ai-fake-image-detector/data/real"
fake_path = "C:/Users/Administrator/ai-fake-image-detector/data/fake"

# Show some real & fake images
real_imgs = os.listdir(real_path)[:3]
fake_imgs = os.listdir(fake_path)[:3]

fig, axes = plt.subplots(2, 3, figsize=(10, 6))
for i, img in enumerate(real_imgs):
    image = cv2.imread(os.path.join(real_path, img))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[0, i].imshow(image)
    axes[0, i].set_title("Real")
    axes[0, i].axis("off")

for i, img in enumerate(fake_imgs):
    image = cv2.imread(os.path.join(fake_path, img))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[1, i].imshow(image)
    axes[1, i].set_title("Fake")
    axes[1, i].axis("off")

plt.show()
```
